```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from statsmodels.formula.api import ols


data = sns.load_dataset("tips")

```

```python
x_tip = data["tip"]
y_total_bill = data["total_bill"]
smoker = data["smoker"]
plt.scatter(x_tip, y_total_bill)
plt.xlabel("Tip")
plt.ylabel("Total Bill")
plt.title("Scatter Plot of\nTip vs Total Bill")
plt.show()
```

```python

num_x = np.size(x_tip)

mean_x = np.mean(x_tip)
mean_y = np.mean(y_total_bill)

cs_d_x = np.sum(y_total_bill*x_tip) - num_x*mean_x*mean_y
cs_d_y = np.sum(x_tip*x_tip) - num_x*mean_x*mean_x

b_1 = cs_d_x/cs_d_y
b_0 = mean_y - b_1*mean_x

coef = (b_1, b_0)
```

```python
predicted_y = b_0 + b_1*x_tip

plt.scatter(x=x_tip,y=y_total_bill)
plt.plot(x_tip, predicted_y, c="red")
plt.show()
```
